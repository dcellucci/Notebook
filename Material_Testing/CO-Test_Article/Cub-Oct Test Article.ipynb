# Analysis for Titanium Cub-Oct Test Coupon

CSL has an opportunity to collaborate with Moog on the performance testing of their new [AM250](http://www.renishaw.com/en/am250-laser-melting-machine--15253) Selective Laser Melting printers. There is one overlap in particular that is interesting- the printing and testing of complex, ultralight geometries that would be prohibitively difficult to construct using traditional methods. 

The first test article is a Cub-Oct lattice. The experiment will use a new Instron machine that will be arriving in December, and will involve tension and compression testing of the sample. 

The intent of this analysis is to ascertain the dimensions of the fixturing elements of the test article, as well as the overall dimensions of the article, in order to allow testing in the Instron. We will use the methods introduced by Ashby in (1), as well as the extensions to these methods that Cheung performed in (2), to find the resulting Young's modulus and ultimate tensile strength (UTS) of the lattice article. We will then use these values to calculate the thickness of the fixturing, by applying precalculated formulae for the deflection of slabs given a distributed load (3).

Initial prototypes of the lattice suggest that the minimum comfortable strut dimensions (at this point) are rectangular, with a width of $0.5\ mm$ and a length of $6.57\ mm$. This produces an octahedral element that fits within a $10\ mm$ cube and has a material volume of $19.8\ mm^3$.

The relative density of the lattice compared to solid titanium is therefore $\rho^*/\rho = 0.0198$

The datasheet for the Ti6Al4V powder, which the AM250 uses, lists the the density as $4420\ kg/m^3$. Additionally, after heat-treating a sample, the modulus of elasticity is $116\ GPa$, and the ultimate tensile strength is $1133\ MPa$.

## Modulus

Assuming that the Cub-Oct structure behaves in the manner derived by Cheung in (2), the resulting modulus for the lattice should be 

$$
E_{lat} = \left(\frac{\rho^*}{\rho}\right)^{1.5} E_{solid}
$$

or about $323\ MPa$

```python
#Properties of AM250 Ti6Al4V 

density = 4420 #kg/m^3
modulus = 116e9 #Pa
uts = 1133e6 #Pa
poisson = 0.33

#Properties of the lattice
voxel_dim = 10 # mm on each side

rel_den = 0.0198

lat_den = rel_den*density #kg/m^3
lat_uts = rel_den**1.5*uts #Pa
lat_mod = rel_den**1.5*modulus #Pa
```

```python
title("Modulus versus Density for AM250 Ti6Al4V")

xscale('log')
yscale('log')
xlim([10,1e4])
ylim([1e5,1e12])
ylabel('Modulus (Pa)')
xlabel('Density (kg/m^3)')

plot(density,modulus,'ro',label='Solid')
plot([1,density],[1.0/density*modulus,modulus],'r--')
plot([1,density],[(1.0/density)**2*modulus,modulus],'r--')
fill_between([1,density],[1.0/density*modulus,modulus],[(1.0/density)**2*modulus,modulus],facecolor='red',alpha=0.1)
plot([1,density],[(1.0/density)**1.5*modulus,modulus],'b--')
plot(lat_den,lat_mod,'bo')

text(1.1e1,3e10,r'$\Delta = 1$ (Ideal Stretch-Dominated Behavior)',size=10,rotation=15)
text(1.1e1,6e7,r'$\Delta = 1.5$',size=10,rotation=22.5)
text(1.1e1,5e8,r'$\Delta = 2$ (Ideal Bending-Dominated Behavior)',size=10, rotation=30)

red_patch = mpatches.Patch(color='red',alpha=0.1, label='Performance Envelope')
rd = mlines.Line2D([], [], color='r', marker='o', linewidth=0, label='Solid')
bd = mlines.Line2D([], [], color='b', marker='o', linewidth=0, label='Lattice')
legend(handles=[red_patch,rd,bd],loc=4, numpoints=1)

show()
```

## Ultimate Tensile Strength

The next calculation is the ultimate tensile strength of the lattice. This also follows the 1.5 power law, and is therefore about 3.156 MPa.

```python
title("Ultimate Tensile Strength versus Density for AM250 Ti6Al4V")

xscale('log')
yscale('log')
xlim([10,1e4])
ylim([1e5,1e10])
ylabel('Ultimate Tensile Strength (Pa)')
xlabel('Density (kg/m^3)')

plot(density,uts,'ro')
plot([1,density],[(1.0/density)**1.5*uts,uts],'b--')

plot(lat_den,lat_uts,'bo')

text(1.1e1,5e5,r'$\Delta = 1.5$',size=10,rotation=30)

rd = mlines.Line2D([], [], color='r', marker='o', linewidth=0, label='Solid')
bd = mlines.Line2D([], [], color='b', marker='o', linewidth=0, label='Lattice')
legend(handles=[rd,bd],loc=4, numpoints=1)

show()
```

## Expected Forces

The proposed overall dimensions of the test coupon have to take into account the minimum number of lattice elements in order for the structure to be considered a cellular solid (as opposed to a truss). The numbers vary, though Ansys simulations have shown convergence for volumes with dimensions exceeding 6 cells in any direction.

As a result, our initial overall dimensions are 16 cells in both horizontal directions, and 8 voxels in the vertical. This produces a cross-sectional area of the sample of $0.0256\ m^2$, and a height of $80\ mm$.

The resulting load-displacement curve is shown below. Failure will occur at a load of $80\ kN$, with a displacement of $0.78\ mm$.

```python
lat_ult_strain = lat_uts/lat_mod

horiz_voxels = 10
ver_voxels = 16

length = horiz_voxels*voxel_dim #mm
area = (length/1000.0)**2 #m^2
height = voxel_dim*ver_voxels #mm

xlim([0,1.1e-2*height])
ylim([0,3.5e6*area])

ylabel('Load(N)')
xlabel('Displacement(mm)')

plot([0,lat_ult_strain*height],[0,lat_uts*area])
title("Load-Displacement for TI4Al6V, {0}x{0}x{1} Voxel Sample".format(horiz_voxels, ver_voxels))

plot(lat_ult_strain*height,lat_uts*area,'bx')
gca().get_yaxis().set_major_formatter(plt.LogFormatter(10,  labelOnlyBase=False))

ult_force = lat_uts*area
print("Ultimate Load: {0} N".format(ult_force))
ult_displacement = lat_ult_strain*height

print("Ultimate Displacement: {0} mm".format(ult_displacement))
```

## Fixture Design

Given the ultimate load and displacement we calculated above, we can now calculate the thickness of the plates which will attach the lattice to the testing rig. 

The dimensions are shown below

<img src="https://raw.githubusercontent.com/dcellucci/Notebook/master/Material_Testing/CO-Test_Article/Media/Drawings/COTest.png" width="500px" float="right"/>

The sample will be attached to the instron using 1/4" bolts, spaced 1" apart, over a square border that that is 8" on a side and centered over the sample. The sample will terminate at a half-cell, instead of a full cell, increasing the surface area which connects it to the interface slabs, and insuring that the failure does not occur at this part. 

We can therefore extimate the deflection of the slab as though it were 8" square, clamped on the border, and experiencing an evenly distributed force $P$. The maximum deflection was calculated for this case by Westergaard and is given by (3)

$$
w_{max} = \frac{0.016(1-\nu^2)Pl^4}{Et^3}
$$

Where $\nu$ is the Poisson's ratio for the material, which we will estimate as $0.33$, $P$ is the pressure from the distributed load, $l$ is the width of the slab, $E$ is the Young's modulus, and $t$ is the thickness. 

We can therefore calculate the thickness required to attain a specific maximum deflection by solving for $t$

$$
t = \sqrt[3]{\frac{0.016(1-\nu^2)Pl^4}{Ew_{max}}}
$$

This, of course, will overestimate the necessary thickness by approximately a factor of $(l/a)^{2/3} = 1.18$, where $a$ is the overall width of the sample, but I am not going to correct for it, and in fact going to use it in lieu of a safety factor.   

```python
tar_def = ult_displacement/1000.0/2.0/10.0 #m
slab_length = (int(length/1000.0*39.37)+1.5)*25.4/1000.0
print(slab_length)
#slab_length = 0.203 #m

tar_thickness = (0.016*(1-poisson**2)*lat_uts*slab_length**4/(modulus*tar_def))**(1.0/3.0)

print("Thickness: {:.4} mm".format(tar_thickness*1000))
```

```python
voxels = range(8,17)
hor_thick = []
ver_thick = []
for h_vox in voxels:
    length = h_vox*voxel_dim #mm
    height = 16*voxel_dim
    ult_displacement = lat_ult_strain*height   #mm
    tar_def = ult_displacement/1000.0/2.0/10.0 #m
    slab_length = (int(length/1000.0*39.37)+1.5)*25.4/1000.0
    tar_thickness = (0.016*(1-poisson**2)*lat_uts*(length/1000.0/slab_length)**2*slab_length**4/(modulus*tar_def))**(1.0/3.0)
    hor_thick.append(tar_thickness*1000)

for v_vox in voxels:
    length = 8*voxel_dim #mm
    height = v_vox*voxel_dim
    ult_displacement = lat_ult_strain*height   #mm
    tar_def = ult_displacement/1000.0/2.0/10.0 #m
    slab_length = (int(length/1000.0*39.37)+1.5)*25.4/1000.0
    tar_thickness = (0.016*(1-poisson**2)*lat_uts*(length/1000.0/slab_length)**2*slab_length**4/(modulus*tar_def))**(1.0/3.0)
    ver_thick.append(tar_thickness*1000)
    
height = voxel_dim*ver_voxels

# Two subplots, the axes array is 1-d
f, axarr = plt.subplots(2, sharex=True)
axarr[0].scatter(voxels, hor_thick)
ylabel('Thickness (mm)')
axarr[0].set_title('Fixture Thickness NxNx16 Article')
axarr[1].set_title('Fixture Thickness 8x8xN Article')
axarr[1].scatter(voxels, ver_thick)
```

## Ultralight

Even at the most compliant article configuration (8x8x16), the resulting fixture thickness is, unfortunately, past what the manufacturer is comfortable making as a flat slab. Our solution was therefore to double the size of the individual voxel while keeping the strut cross section constant.

The subsequent lattice has a material volume of $41.02\ mm^3$ per voxel, which fills a volume of $8000\ mm^3$. The resulting relative density is therefore $\rho^*/\rho = 0.00513$.

The question is, then, does this lower relative density result in a substantively different fixture thickness.

```python
#Properties of the ultralight lattice
ul_voxel_dim = 20 # mm on each side

ul_rel_den = 41.02/8000

ul_lat_den = ul_rel_den*density #kg/m^3
ul_lat_uts = ul_rel_den**1.5*uts #Pa
ul_lat_mod = ul_rel_den**1.5*modulus #Pa
```

```python
ul_lat_ult_strain = ul_lat_uts/ul_lat_mod

horiz_voxels = 8
ver_voxels = 8

length = horiz_voxels*ul_voxel_dim #mm
area = (length/1000.0)**2 #m^2
height = ul_voxel_dim*ver_voxels #mm

xlim([0,1.1e-2*height])
ylim([0,5e5*area])

ylabel('Load(N)')
xlabel('Displacement(mm)')

plot([0,ul_lat_ult_strain*height],[0,ul_lat_uts*area])
title("Load-Displacement for TI4Al6V, {0}x{0}x{1} Voxel Sample".format(horiz_voxels, ver_voxels))

plot(ul_lat_ult_strain*height,ul_lat_uts*area,'bx')
gca().get_yaxis().set_major_formatter(plt.LogFormatter(10,  labelOnlyBase=False))

ul_ult_force = ul_lat_uts*area
print("Ultimate Load: {0} N".format(ul_ult_force))
ul_ult_displacement = ul_lat_ult_strain*height

print("Ultimate Displacement: {0} mm".format(ul_ult_displacement))

tar_def = ul_ult_displacement/1000.0/2.0/10.0 #m
slab_length = (int(length/1000.0*39.37)+1.5)*25.4/1000.0
print(slab_length)
#slab_length = 0.203 #m

tar_thickness = (0.016*(1-poisson**2)*ul_lat_uts*slab_length**4/(modulus*tar_def))**(1.0/3.0)

print("Thickness: {:.4} mm".format(tar_thickness*1000))
```

The resulting thickness is still $9.516\ mm$, which is somewhat odd. 

We can derive a formula for $t$ that depends only on the number of voxels, the octahedron volume, and the length of the cube that is the voxel bounding box. Recall


$$
t = \sqrt[3]{\frac{0.016(1-\nu^2)Pl^4}{Ew_{max}}}
$$

if we substitute $\alpha = 0.016(1-\nu^2)$, and note that

$$
P = U_{Ti}\left(\frac{v_{oct}}{v_{cub}}\right)^{3/2} = U_{Ti}\left(\frac{v_{oct}}{l_{cub}^3}\right)^{3/2},
$$

where $U_{Ti}$ is the ultimate tensile strength of the material,

$$
w_{max} = \frac{U_{ti}}{E}ml_{cub},
$$

where $m$ is the number of vertical voxels, and finally

$$
l = nl_{cub},
$$

where $n$ is the number of horizontal voxels

The resulting required thickness is therefore

$$
t \propto \sqrt[3]{\frac{U_{Ti}\left(\frac{v_{oct}}{l_{cub}^3}\right)^{3/2}(nl_{cub})^4}{E\frac{U_{ti}}{E}ml_{cub}}} 
$$

$$
t \propto \sqrt[3]{\frac{n^4}{m}\left(\frac{v_{oct}}{l_{cub}}\right)^{3/2}} 
$$

As we can see, doubling the voxel size does nothing, since doubling the strut length doubles both the octahedron volume and the length of the bounding box for the voxel.
