```python
%pylab inline
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
import matplotlib.pyplot as plt
```

# Analysis for Titanium Cub-Oct Test Coupon

Octahedron volume is $19.8\ mm^3$

Cell volume is $1000\ mm^3$

Relative density is therefore $\rho^*/\rho = 0.0198$

Modulus of Elasticity: $116\ GPa$

```python
xscale('log')
yscale('log')
xlim([10,1e4])
ylim([1e5,1e12])
ylabel('Modulus (Pa)')
xlabel('Density (kg/m^3)')
plot(4420,116e9,'ro',label='Solid')
plot([1,4420],[26.2e6,116e9],'r--')
plot([1,4420],[5.937e3,116e9],'r--')
fill_between([1,4420],[26.2e6,116e9],[5.937e3,116e9],facecolor='red',alpha=0.1)

plot([1,4420],[3.94e5,116e9],'b--')
plot(87.51,323e6,'bo')
text(1.1e1,3e10,r'$\Delta = 1$ (Ideal Stretch-Dominated Behavior)',size=10,rotation=15)
text(1.1e1,6e7,r'$\Delta = 1.5$',size=10,rotation=22.5)
text(1.1e1,5e8,r'$\Delta = 2$ (Ideal Bending-Dominated Behavior)',size=10, rotation=30)

red_patch = mpatches.Patch(color='red',alpha=0.1, label='Performance Envelope')
rd = mlines.Line2D([], [], color='r', marker='o', linewidth=0, label='Solid')
bd = mlines.Line2D([], [], color='b', marker='o', linewidth=0, label='Experiment')
legend(handles=[red_patch,rd,bd],loc=4, numpoints=1)

show()
```

```python
xscale('log')
yscale('log')
xlim([10,1e4])
ylim([1e5,1e10])
ylabel('Yield Strength (Pa)')
xlabel('Density (kg/m^3)')
plot(4420,1133e6,'ro')
plot([1,4420],[3.855e3,1133e6],'b--')

plot(87.51,3.156e6,'bo')

text(1.1e1,5e5,r'$\Delta = 1.5$',size=10,rotation=30)

rd = mlines.Line2D([], [], color='r', marker='o', linewidth=0, label='Solid')
bd = mlines.Line2D([], [], color='b', marker='o', linewidth=0, label='Experiment')
legend(handles=[rd,bd],loc=4, numpoints=1)
show()
```

```python
xlim([0,1.1e-2])
ylim([0,3.5e6])

ylabel('Stress(Pa)')
xlabel('Strain')

plot([0,9.77e-3],[0,3.156e6])

plot(9.77e-3,3.156e6,'bx')

gca().get_yaxis().set_major_formatter(plt.LogFormatter(10,  labelOnlyBase=False))

show()
```

```python
horiz_voxels = 16
ver_voxels = 8
area = (horiz_voxels*10.0/1000.0)**2 #m^2
height = 10*ver_voxels #mm

xlim([0,1.1e-2*height])
ylim([0,3.5e6*area])

ylabel('Load(N)')
xlabel('Displacement(mm)')

plot([0,9.77e-3*height],[0,3.156e6*area])


plot(9.77e-3*height,3.156e6*area,'bx')
gca().get_yaxis().set_major_formatter(plt.LogFormatter(10,  labelOnlyBase=False))

ult_force = 3.156e6*area
print(ult_force)
```
